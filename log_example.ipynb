<p>
자연로그(Natural Logarithm, ln)란?
   <br/>Base가 e (오일러상수 2.718...)인 로그, Base인 e에 몇 승을 해야 입력한 숫자가 나오는지 계산

   - ln 1       = 0.0
   - ln 2       = 0.69
   - ln 2.718   = 1.0    => e의 1승은 자신인 2.718... 나오기 때문
   - ln 7.389   = 2.0    => e의 2승은 e의 제곱인 7.389... 나오기 때문
   - ln 20.085 = 3.0</p>

<p>
상용로그(Common Logarithm)란?
    <br>Base가 10인 로그, Base인 10에 몇 승을 해야 입력한 숫자가 나오는지 계산

    - log10 1    = 0.0
    - log10 2    = 0.3
    - log10 10   = 1.0    => 10의 1승은 자신인 10이 나오기 때문
    - log10 100  = 2.0    => 10의 2승은 100이 나오기 때문
    - log10 1000 = 3.0</p>

```python
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# X값 생성
x = np.linspace(start=1, stop=10, num=1000)
```

```python
plt.figure(figsize=(10, 6)) # 크기 지정

# 자연로그(Natural Logarithm) Y값 생성 
plt.plot(x, np.log(x), 'b-', label='ln')

# 상용로그(Common Logarithm) Y값 생성 
plt.plot(x, np.log10(x), 'r-', label='log10')

# 밑이 2인 로그 Y값 생성 
plt.plot(x, np.log2(x), 'k-', label='log2')

plt.legend() # 범례출력
```

로그함수(logarithmic function)는 지수함수의 역함수(inverse function) 
- 지수함수 y=2<sup>x</sup>
- 역함수  x=2<sup>y</sup> 이를 보기 좋게 다시 고쳐 쓴 것이, 다음의 
- 로그함수  y=log<sub>2</sub>x

```python
plt.figure(figsize=(10, 6)) # 크기 지정

plt.plot(x, np.exp(x), 'g-', label='exp')

plt.legend() # 범례출력
```
